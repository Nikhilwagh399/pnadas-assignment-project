```python
## Importing the necessary libraries

import pandas as pd 
```

```python
## importing the dataset

df = pd.read_csv('Data (2).csv')
```

```python
df.head() ## to check the data 
```

```python
df['Date'] = pd.to_datetime({'year': df['Yr'] + 1900, 'month': df['Mo'], 'day': df['Dy']})
```

```python
df['Date'] = df['Date'].apply(lambda x: x.replace(year=x.year - 100) if x.year > 2000 else x)
```

```python
df.set_index('Date', inplace=True) ## Set the new 'Date' column as the index
```

```python
df = df.drop(['Yr', 'Mo', 'Dy'], axis=1) ### Drop the original Yr, Mo, Dy columns
```

```python
df.head()
```

```python
print( df.isnull().sum()) ## # Check for missing values in the dataset

```

```python
df_filled = df.fillna(df.mean()) ##Fill missing values with column
```

```python
print("Missing values after filling:\n", df_filled.isnull().sum()) ## Verify no missing values remain
```

```python
df.describe()
```

```python
loc_stats = pd.DataFrame({          ### Calculate statistics for each location

    'Location': df.columns,
    'Min Windspeed': df.min(),
    'Max Windspeed': df.max(),
    'Mean Windspeed': df.mean(),
    'Std Windspeed': df.std()})
```

```python
##Rename columns for clarity

loc_stats.columns = ['Location', 'Min Windspeed', 'Max Windspeed', 'Mean Windspeed', 'Std Windspeed']
```

```python
print(loc_stats) ### Display the results
```

```python
day_stats = pd.DataFrame({  ##Calculate statistics across all locations for each day
    'Date': df.index,  
    'Min Windspeed': df.min(axis=1),
    'Max Windspeed': df.max(axis=1),
    'Mean Windspeed': df.mean(axis=1),
    'Std Windspeed': df.std(axis=1)})
```

```python
print(day_stats) ## the result
```

```python
df['Month'] = df.index.month  # Extract the month from the index
```

```python
january_df = df[df['Month'] == 1]  # Filter data for January (Month == 1)

```

```python
january_df = january_df.drop(columns=['Month']) # Drop the 'Month' column as it's no longer needed
```

```python
january_means = january_df.mean() ### Calculate the average windspeed for January for each location
```

```python
print("Average windspeed in January for each location:\n",january_means) # Display the resultl
```

```python
yearly_df = df.resample('Y').mean() # Downsample the dataset to a yearly frequency and calculate the mean for each year
```

```python
print("Yearly average windspeed for each location:\n", yearly_df) ### Display the result
```

```python
monthly_df = df.resample('M').mean()# Downsample the dataset to a monthly frequency and calculate the mean for each month
```

```python
print("monthly average windspeed for each location:\n",monthly_df) #show result in monthly 
```

```python
weekly_df = df.resample('W').mean ## as well as on weekly data
```

```python
##Resample data weekly starting from January 2, 1961, and calculate statistics
weekly_stats = df.resample('W-MON', origin='1961-01-02').agg(['min', 'max', 'mean', 'std'])
```

```python
first_52_weeks = weekly_stats.head(52) # Limit to the first 52 weeks
```

```python
print("Weekly statistics (first 52 weeks):\n",first_52_weeks) ##result
```
